#### se hace un diccionario que contenga la region y la tasa


```python
Resultado={}

with open("data/Covid-19.csv") as Covid:
    linea= csv.reader(Covid, delimiter= ",")
    bandera=1
    contador=1
    for columna in linea:
        if bandera ==1:
            for elemento in columna :
                print(elemento)
                bandera=0
        else:
            contador=0
            try:
                    Resultado[columna[0]]["Tasa"]+=float(columna[162])
                
            except KeyError:
                    Resultado[columna[0]]={"Tasa":+float(columna[162])}
                    contador+=1
            
```
